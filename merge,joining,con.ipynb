```python
# Create the first DataFrame (df_left)
df1 = pd.DataFrame({
    'Key': ['K0', 'K1', 'K2', 'K3'],
    'A': ['A0', 'A1', 'A2', 'A3'],
    'B': ['B0', 'B1', 'B2', 'B3']
})

# Create the second DataFrame (df_right)
df2 = pd.DataFrame({
    'Key': ['K0', 'K1', 'K2', 'K3'],
    'C': ['C0', 'C1', 'C2', 'C3'],
    'D': ['D0', 'D1', 'D2', 'D3']
})

# Merge df_left and df_right on the 'Key' column
merged_df = pd.merge(df1, df2, on='Key')


print(df1)

print("\n")

print(df2)
print("\n")


print(merged_df)
```

```python
joined_df=df1.join(df2)
print(df1)
print(df2)
print(joined_df)
```
