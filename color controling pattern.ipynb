```python
import matplotlib.pyplot as plt
x=[1,2,3,4]
y=[2,4,6,8]
plt.plot(x,y,label='line plot')
plt.xlabel('x-axis')
plt.xlabel('y-axis')
plt.title('simple line plot')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4]
y=[2,4,6,8]
plt.plot(x,y,linestyle='--',color='red',marker='0',label='custom line')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Line Plot with Custom Properties')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]
plt.plot(x, y, linestyle='--', color='red', marker='o', label='Custom Line')
plt.xlabel('X-axis') 
plt.ylabel('Y-axis')
plt.title('Line Plot with Custom Properties')
plt.legend()
plt.show()

```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4]
y1=[2,4,6,8]
y2=[1,3,4,5,]
plt.plot(x,y1,linestyle='-',color='blue',label='solid blue line')
plt.plot(x,y2,linestyle='--',color='red',label='dashed red line')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('line plot with different patterns and colors')
plt.legend()
plt.savefig('line_plot.png')
plt.show()
```
