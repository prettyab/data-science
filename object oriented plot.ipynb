```python
import matplotlib.pyplot as plt
subjects=['maths','chemistry','physics']
mark=[30,45,60]
plt.bar(subjects,mark,color=['skyblue','lightgreen','orange'])
plt.xlabel('subjects')
plt.ylabel('mark')
plt.title('simple bar chart')
plt.show()
```

```python
import matplotlib.pyplot as plt
fruits=['apple','orange','grape']
rate=[120,234,210]
plt.bar(fruits,rate,color=['red','blue','green'])
plt.xlabel('fruits')
plt.ylabel('rate')
plt.title('simple bar chart')
plt.show()
```

```python
import matplotlib.pyplot as plt
labels=['catrgory A','category B','category C']
sizes=[20,30,20]
plt.pie(sizes,labels=labels,autopct='%1.1f%%',startangle=90,colors=['skyblue','red','green'])
plt.title('simple pie chart')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
data=np.random.randn(1000)
plt.hist(data,bins=25,edgecolor='black')
plt.xlabel('value')
plt.ylabel('Frequency')
plt.title('histogram graph')
plt.show()

```

```python
import matplotlib.pyplot
import numpy as np
np.random.seed(42)
data=[np.random.normal(0,std,100)for std in range(1,4)]
plt.boxplot(data,labels=['product1','product2','product3'],notch=True,vert=True,patch_artist=True)
plt.xlabel('products')
plt.ylabel('values')
plt.title('Boxplot graph')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(42)
x=np.random.rand(30)*10
y=2*x+1+np.random.randn(30)
plt.scatter(x,y,color='blue',marker='o')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('simple scatter plot')
plt.show()
```
